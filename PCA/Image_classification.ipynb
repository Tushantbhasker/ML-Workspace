# First we apply PCA to our dataset to reduce the no. of features 
- It will reduce the accuracy a little bit.
- But also reduce the the computational time of our classifier
- We will apply the PCA at 99% of variance.

```python
import sklearn.datasets
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import ensemble
import time
```

```python
dataset = sklearn.datasets.fetch_lfw_people(min_faces_per_person = 100, resize = 0.4)
```

```python
print(dataset.keys())
print(dataset.data.shape)
print(dataset.images.shape)
print(dataset.target.shape)
print(dataset.target_names)
x = dataset.data
y = dataset.target
images = dataset.images
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 0)
x_train.shape
```

```python
# image_train = x_train.resize((855,50,37))
# print(type(image_train))
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(images[i], cmap = plt.cm.bone)
plt.show()
```

```python
pca = PCA()
pca.fit(x_train)
```

```python
# finding  the optimal value of no. of features (k)
k = 0
total = sum(pca.explained_variance_)
current_sum = 0
while current_sum/total < 0.99:
    current_sum += pca.explained_variance_[k]
    k += 1
k
```

```python
pca = PCA(n_components = k, whiten = True)
x_train_t = pca.fit_transform(x_train)
```

```python
x_train_t.shape
```

```python
x_approx = pca.inverse_transform(x_train_t)
x_approx.shape
x_approx_images = x_approx.reshape((855, 50 ,37))
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(x_approx_images[i], cmap = plt.cm.bone)
plt.show()
```

```python
# printing igenfaces 
eigenValue = pca.components_
print(eigenValue.shape)
eigenValue = eigenValue.reshape((316, 50 ,37))
eigenValue.shape
```

```python
fig = plt.figure(figsize = (8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(eigenValue[i], cmap = plt.cm.bone)
plt.show()
```

```python
x_test_t = pca.transform(x_test)
x_test_t.shape
```

```python
# without PCA
clf = ensemble.RandomForestClassifier()
start = time.time()
clf.fit(x_train,y_train)
print(time.time() - start)
y_pred = clf.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
clf = ensemble.RandomForestClassifier()
start = time.time()
clf.fit(x_train_t,y_train)
print(time.time() - start)
y_pred = clf.predict(x_test_t)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
