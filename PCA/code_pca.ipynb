```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243)
```

```python
mean_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1, cov_mat1, 100)
```

```python
mean_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2, cov_mat2, 100)
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, proj3d

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.plot(class1[:, 0], class1[:, 1], class1[:, 2], 'o')
ax.plot(class2[:, 0], class2[:, 1], class2[:, 2], '^')
plt.show()
```

```python
all_data = np.concatenate((class1, class2))
```

```python
pca = PCA(n_components = 2)
transformed_data = pca.fit_transform(all_data)
transformed_data
```

```python
# Eigen vector claculated by the in built pca in sklearn
pca.components_
```

```python
# Our own pca code
all_data_t = all_data.T #taking transpose of data
cov = np.cov(all_data_t) #taking ovariance of data feature wise 
cov
```

```python
eigen_value, eigen_vector = np.linalg.eig(cov)
eigen_value, eigen_vector  #eigen_vector are column wise
```
