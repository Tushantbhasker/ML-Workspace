```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\tushant\Desktop\ML\LogisticRegression\titanic\training_titanic_x_y_train.csv")
data.head()
```

```python
df = data.copy()
df.drop("Name", axis=1, inplace=True)
df.drop("Cabin", axis=1, inplace=True)
df.head()
```

```python
def f(s):
    if s == "male":
        return 1
    else:
        return 0
def f1(s):
    if s == "S":
        return 0
    elif s == "C":
        return 1
    elif s == "Q":
        return 2
    else:
        return 0

df["Gender"] = df.Sex.apply(f)
df.drop("Sex",axis=1, inplace=True)
df["E"] = df.Embarked.apply(f1)
df.drop("Embarked", axis=1, inplace=True)
df.drop("Ticket", axis=1, inplace=True)
df.head()
```

```python
a = df.Age.mean()
a = a// 1
df["Age"].fillna(a,inplace=True)
df.head()
```

```python
y = df.values[:,5]
df.drop("Survived",axis=1,inplace=True)
# print(df.head())
x = df.values[:,1:]
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
alg = LogisticRegression()
```

```python
alg.fit(x,y)
```

```python
x_test = pd.read_csv(r"C:\Users\tushant\Desktop\ML\LogisticRegression\titanic\test_titanic_x_test.csv")
x_test["Gender"] = x_test.Sex.apply(f)
x_test.drop("Sex",axis=1, inplace=True)
x_test["E"] = x_test.Embarked.apply(f1)
x_test.drop("Embarked", axis=1, inplace=True)
x_test.drop("Ticket", axis=1, inplace=True)
x_test.drop("Name", axis=1, inplace=True)
x_test.drop("Cabin", axis=1, inplace=True)
x_test["Age"].fillna(a,inplace=True)
x_test = x_test.values[:,1:]
x_test.shape
```

```python
y_pre= alg.predict(x_test)
np.savetxt("text.csv",y_pre)
y_pre.shape
```
