```python
x = data[:,0:4]
a = np.ones(len(x),int).reshape(-1,1)
x = np.hstack((x,a))
y = data[:,4]
print(x.shape)
print(y.shape)
```

```python
def cost(x, y, m):
    N = len(x)
    cost = 0
    for i in range(N):
        y_p = 0
        for j in range(5):
            y_p += x[i][j] * m[j] 
        cost += (1/N) * (y[i] - y_p) ** 2
    return cost
```

```python
def step_gradient(x, y, k, m):
 
    m_slope = np.zeros(5)
    N = len(x)
    for i in range(5):
        for j in range(N):
            y_p =(m * x[j]).sum()
            m_slope[i] += (-2/N) * (y[j] - y_p ) * x[j][i] 
           
    m_new = m - k * m_slope
#     print(m_new)
    return m_new
            
```

```python
def gradient_descent(x, y, k, n):
    m = np.zeros(5)
    for i in range(n):
        m = step_gradient(x, y, k, m)
        print(i,"Cost:",cost(x,y,m))
    return m
```

```python
def run(x, y):
    learning_rate = 0.00000095
    number_iterations = 7000
    m = gradient_descent(x, y, learning_rate, number_iterations)
    return m
```

```python
m = run(x,y)
```
